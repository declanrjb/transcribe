```python
from dotenv import load_dotenv
import json
import whisper
```

```python
load_dotenv('openai.env')
```

```python
model = whisper.load_model("base")
result = model.transcribe("audio_sample.m4a")
```

```python
with open('transcript.json', 'w') as out_path:  
    json.dump(result, out_path)
```
